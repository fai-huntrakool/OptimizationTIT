```python
import numpy as np
import matplotlib.pyplot as plt
import math 
import seaborn as sns
```

```python
plt.style.use('seaborn')
```

```python
#exponential distribution
def p_e(x):
    if x < 0:
        return 0
    else:
        return 0.5*np.exp(-0.5*x)*1/math.factorial(5)*x**4*np.exp(-x)
```

```python
#gamma distribution
def p_g(x):
    if x <= 0:
        return 0
    else:
        return 1/math.factorial(5)*(x**4)*np.exp(-x)
```

```python
def p(x):
    return p_g(x)*p_e(x)
```

```python
size = [1000, 5000, 10000, 20000, 30000,100000]
color = ['r','y','b','g','m','k']
```

```python
#Metropolis-Hastings
count = 0
for n in size:
    x = np.arange(0.0,n*1.0)
    x[0] = 3.0
    for i in range(1,n):
        currentx = x[i-1]
        proposedx = currentx + np.random.normal(0,1,1) #mean,std,size
        A = p(proposedx)/p(currentx)
        if np.random.rand() <= A:
            x[i] = proposedx
        else:
            x[i] = currentx
    sns.distplot(x, label = 'iteration '+str(n), color=color[count], hist = False)
    count+=1
plt.xlabel("x")
plt.ylabel("Distribution")
plt.title("Distribution of X")
```
