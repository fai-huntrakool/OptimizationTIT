```python
import numpy as np
import matplotlib.pyplot as plt
import math 
import seaborn as sns
```

```python
plt.style.use('seaborn')
```

```python
def p_e(x):
    if x <= 0:
        return 0
    else:
        return (0.5*np.exp(-0.5*x)) * (1/math.factorial(5)*(x**4)*np.exp(-x))
```

```python
size = [1000, 5000, 10000, 20000, 30000]
color = ['r','y','b','g','m']
```

```python
#Metropolis-Hastings
count = 0
for n in size:
    x = np.arange(0.0,n*1.0)
    x[0] = 3.0
    for i in range(1,n):
        currentx = x[i-1]
        proposedx = currentx + np.random.normal(0,1,1) #mean,std,size
        A = p_e(proposedx)/p_e(currentx)
        if np.random.rand() <= A:
            x[i] = proposedx
        else:
            x[i] = currentx
    sns.distplot(x, label = 'iteration '+str(n), color=color[count], hist = False)
    count+=1
plt.xlabel("x")
plt.ylabel("Distribution")
plt.title("Distribution of X")
```
